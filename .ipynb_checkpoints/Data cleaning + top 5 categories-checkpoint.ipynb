```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
content = pd.read_csv('C://Users//julia//Documents//Projects DS//Accenture Data Visualization and Storytelling//Content.csv')
reactions = pd.read_csv('C://Users//julia//Documents//Projects DS//Accenture Data Visualization and Storytelling//Reactions.csv')
reaction_types = pd.read_csv('C://Users//julia//Documents//Projects DS//Accenture Data Visualization and Storytelling//ReactionTypes.csv')
```

```python
content.head()
```

```python
reactions.head()
```

```python
reaction_types.head()
```

```python
content = content.drop('URL',axis=1)
content
```

```python
content['Category'].unique()
```

```python
content['Category'].replace('"','',inplace=True,regex=True)
content
```

```python
content['Category'].unique()
```

```python
content['Category'] = content['Category'].str.strip()
content
```

```python
content['Category'] = content['Category'].apply(lambda x: x.capitalize())
content
```

```python
content['Type'] = content['Type'].apply(lambda x: x.capitalize())
content
```

```python
content.rename(columns={'Type':'Content type'},inplace=True)
content
```

```python
reactions.rename(columns={'Type': 'Reaction type'},inplace=True)
reactions
```

```python
reaction_types.rename(columns={'Type': 'Reaction type'},inplace=True)
reaction_types
```

```python
content = content.drop('User ID',axis=1)
content
```

```python
reactions = reactions.drop('User ID',axis=1)
reactions
```

```python
reactions = reactions.dropna(axis=0,subset=['Reaction type'])
```

```python
reactions
```

```python
reactions.isna().sum()
```

```python
content = content.drop('Unnamed: 0',axis=1)
content
```

```python
reactions = reactions.drop('Unnamed: 0',axis=1)
reactions
```

```python
reaction_types = reaction_types.drop('Unnamed: 0',axis=1)
reaction_types
```

```python
temp_df = pd.merge(content,reactions,on='Content ID')
df = temp_df.merge(reaction_types,on='Reaction type',how='left')
df
```

```python
categories = df.groupby(['Category']).sum('Score').sort_values('Score',ascending=False)
categories
```

```python
categories = categories.reset_index()
categories
```

```python
top5 = categories[:5]
top5
```

```python
fig,ax = plt.subplots(figsize=(16,8))
sns.barplot(data=top5,x='Category',y='Score')
plt.suptitle('Top 5 categories')
plt.ylim(65000,76000)
plt.show()
```

```python
categories.to_csv('Categories.csv')
```

```python
df.to_csv('Final dataset.csv')
```
